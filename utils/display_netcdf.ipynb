```python
import netCDF4 as nc
import matplotlib.pyplot as plt
import numpy as np


```

```python

num_Files = 10
plt.figure(figsize=(15, 10))  # Adjust the figure size as necessary
for i in range(num_Files):
    # Load the NetCDF file
    # file_path = f"../simulation/app/data/split/hydrodynamic_W_d000{i}.nc"  # Replace this with the actual path to your NetCDF file
    file_path = f"../simulation/app/data/perlin/daily_netcdf/hydrodynamic_W_d000{i}.nc"  # Replace this with the actual path to your NetCDF file
    dataset = nc.Dataset(file_path)

    # Extract the variable 'W'
    w_variable = dataset.variables['w']

    d = 16
    # Create plots for each time step
    plt.subplot(int(np.ceil(np.sqrt(num_Files))), int(np.ceil(np.sqrt(num_Files))), i + 1)
    plt.imshow(w_variable[0, d, :, :], cmap='viridis', vmin=-0.001, vmax=0.001)
    plt.colorbar()
    plt.title(f'W at time step {i}, depth {d}')
    plt.xlabel('X dimension')
    plt.ylabel('Y dimension')
plt.tight_layout()
plt.show()


```
